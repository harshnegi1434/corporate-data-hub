```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load the dataset into a DataFrame
df = pd.read_csv('company_data.csv')
```

```python
df.tail()
```

```python
num_records = len(df)
print("Number of records:", num_records)
```

```python
df.shape
```

```python
df.columns
```

```python
filtered_df = df[df['Company Name'] == 'Microsoft']
print(filtered_df)
```

```python
# Removing non-required symbols in Salary Column
df['Salary'] = df['Salary'].str.replace('₹', '').str.replace(',', '').astype(int)
```

```python
df['Salary'].head()
```

```python
# Converting Salary which are given in monthly format (5 digit) into yearly format
df.loc[df['Salary'].astype(str).str.len() == 5, 'Salary'] *= 12
```

```python
(df['Salary'].astype(str).str.len() == 5).sum()
```

```python
# Converting now yearly salary into LPA format
df['Salary'] = df['Salary'] / 100000
df['Salary'] = df['Salary'].round(2)  
```

```python
df['Salary'].head(10)
```

```python
# Removing (',') symbol from each value in the 'Reported Salaries' column
df['Reported Salaries'] = df['Reported Salaries'].str.replace(',', '')
# Removing the word 'salaries' from each value in the 'Reported Salaries' column
df['Reported Salaries'] = df['Reported Salaries'].str.replace(' salaries', '')

df['Reported Salaries'] = pd.to_numeric(df['Reported Salaries'])
```

```python
df['Reported Salaries'].head()
```

```python
# Shuffling the records
df = df.sample(frac=1, random_state=42) 
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.to_csv('company_data_processed.csv', index=False)
```

```python
salary_bins = [0, 4, 6, 8, 10, 15, 20, 30, float('inf')]
salary_labels = ['0-4', '4-6', '6-8', '8-10', '10-15', '15-20', '20-30', '30+']

df['Salary Range'] = pd.cut(df['Salary'], bins=salary_bins, labels=salary_labels, right=False)

salary_range_counts = df['Salary Range'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
salary_range_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Companies by Salary Range')
plt.xlabel('Salary Range (LPA)')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
rating_counts = df['Rating'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
plt.bar(rating_counts.index, rating_counts.values, color='blue')
plt.xlabel('Rating')
plt.ylabel('Count of Companies')
plt.title('Distribution of Companies by Rating')
plt.show()
```

```python
text = ' '.join(df['Company Name'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Salary', data=df)
plt.title('Summary Statistics of Salaries')
plt.show()
plt.savefig('summary_sns.png')
```

```python
designation_counts = df['Designation'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(designation_counts, labels=designation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Each Designation Category')
plt.show()
```

```python
plt.scatter(df['Salary'], df['Rating'], color='green', alpha=0.5)
plt.xlabel('Salary')
plt.ylabel('Rating')
plt.title('Relationship between Salary and Rating')
plt.show()
```

```python
corr_matrix = df[['Salary', 'Rating', 'Reported Salaries']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
idx_max_salary = df.groupby('Designation')['Salary'].idxmax()
highest_salary_company = df.loc[idx_max_salary, ['Designation', 'Company Name', 'Salary']]
highest_salary_company.reset_index(drop=True, inplace=True)
print("\nHighest salary and corresponding company name for each designation (in LPA):")
print(highest_salary_company)
```

```python
idx_max_rating = df.groupby('Designation')['Rating'].idxmax()
highest_rating_company = df.loc[idx_max_rating, ['Designation', 'Company Name', 'Rating']]
highest_rating_company.reset_index(drop=True, inplace=True)
highest_rating_company.sort_values(by='Designation', inplace=True)
print("\nHighest rating and corresponding company name for each designation:")
print(highest_rating_company)
```

```python
idx = df.groupby('Designation')['Reported Salaries'].transform("max") == df['Reported Salaries']
highest_reported_salaries = df[idx][['Designation', 'Company Name', 'Reported Salaries']]
highest_reported_salaries.reset_index(drop=True, inplace=True)
print("\nCompany name with the highest reported salaries for each designation:")
print(highest_reported_salaries)
```

```python
grouped_df = df.groupby('Company Name')['Reported Salaries'].sum().reset_index()
sorted_df = grouped_df.sort_values(by='Reported Salaries', ascending=False)
top_5_reported_salaries_company = sorted_df.head(5)
top_5_reported_salaries_company.reset_index(drop=True, inplace=True)
print("\nTop 5 companies with the highest reported salaries (aggregated across all designations):")
print(top_5_reported_salaries_company)
```
