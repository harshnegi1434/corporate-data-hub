```python
from selenium import webdriver
driver = webdriver.Chrome()
url = "https://www.glassdoor.co.in/Salaries/software-engineer-salary-SRCH_KO0,17.htm"
driver.get(url)
driver.implicitly_wait(10)
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
driver.quit()
```

```python
sde_companies_name = []
for i in soup.findAll(class_="css-pmokl e1aj7ssy3"):
    sde_companies_name.append(i.text)
```

```python
sde_companies_name
```

```python
sde_companies_salary = []
for i in soup.findAll('h3', class_="m-0 css-16zrpia el6ke054"):
    salary_text = i.text.strip()
    if salary_text.startswith('₹'):
        sde_companies_salary.append(salary_text)
```

```python
sde_companies_salary
```

```python
sde_companies_ratings = []
for i in soup.findAll('span', class_="m-0 css-60s9ld el6ke050"):
    sde_companies_ratings.append(i.text)
```
